```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# dummy

data = {
    'text' : [
        'gratis uang tunai',
        'diskon besar di shoope',
        'meeting jam 9 pagi',
        'bring it home',
        'hadiah gratis menanti',
        'proyek ai is done'
    ],
    'label' : ['spam','spam,','ham', 'ham', 'spam', 'ham']
}

df = pd.DataFrame(data)
print(df)

# prepocesing : mengubah teks menjadi angka ( vektorisasi )
# komputer tidak bisa baca teks , so kita harus ubah jadi matrix

vectorizer = CountVectorizer
X_train = vectorizer.fit_transform(df['text'])
y_train = df['label']

# tarinng model ( inilah proses menghitung prior dan lieklilhood
# kita pake multinominalnb karena fitur kita ada frekuensi kata (diskrit)

model = MultinomialNB()
model.fit(X_train, y_train)

print("model berhasil dilatih dan mesin sudah menghitung probabilitasnya ")

#4 prediksi data baru
email_baru  = [
    "klik link untuk haiah",
    "neosk kita meet"
]

# ubah data baru ke fpormat angka uang sama dengan trainin g
X_test = vectorizer.transform(email_baru)

# prediksi
prediksi = model.predict(X_test)
probabilitas = model.predict(X_test)

# hasil

for i, text in enumerate(email_baru):
    kelas = prediksi[i]
    # prpbabilitas peluang ham san spam
    confidence = np.max(probabilitas[i]) * 100

    print(f"Email: '{text}'")
    print(f"Prediksi: {kelas.upper()} (Yakin {confidence:.2f}%)")
    print("-" * 30)
```
