# NORMALISASI DATA ( TEKNIK UMUM ML )

```python
data = np.array([[1,2],[3,4],[5,6]])
data
```

```python
mean = np.mean(data, axis=0)
mean
```

```python
std = np.std(data, axis=0)
std
```

```python
normalized = (data - mean) / std # Broadcasting
normalized
```
