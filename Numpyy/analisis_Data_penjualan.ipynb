## LOADING DATA DENGAN NUMPY

```python
# BACA CSV MENGGUNAKNA PYTHON  BUILT-IN UNTUK HANDLER HEADER
with open('C:/Users/test/OneDrive/Desktop/PYTHON/dataset/SuperMarketAnalysis.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader) # skip header
    data = list(reader)

print(f"DATA BERHASIL DIMUAT : {len(data)} baris")
print(f"Kolom : {','.join(header)}")

```

# MENAMPILKAN HEADER DENGAN INDEX

```python
print("STUKRUR KOLOM : ")
for i, col in enumerate(header):
    print(f"    [{i:2d} {col}")
```

# PREVIEW 5 BARIS PERTAMA

```python
for i, row in enumerate(data[5],1):
    print(f"\n--- BARIS {i} ----")
    for j, (col_name, value) in enumerate(zip(header,row)):
        # format khushs unutk kolom numerik
        if j >=6 and j!= 10 and j != 11 and j!= 12: # ski[ date, time , payment
            try:
                val_float = float(value)
                print(f"    {col_name:25} : {val_float: >10.2f}")
            except:
                print(f"    {col_name:25} : {value}")
            else:
                print(f"    {col_name:25} : {value}")
```

# info tambahan

```python
print("INFO DATASET")
print()
print(f"    TOTAL TRANSAKSI     :   {len(data)}")
print(f"    KOLOM               : {','.join(header)}")
print(f"    KOLOM NUMERIK       :   Unit Price, Quantity, Tax 5%, Total, cogs, gross income, Rating")
print(f"    Branch, City, Customer type, Gender, Product line, Payment")
print(f"    Kolom Datetime      : Date, Time")
```

# konversi ke numpy arrays

```python
# konversi ke numpy array untuk kolom numerik yang dibutuhkan
# kolom : unit price(6), quantity(7), tax(8), total(9), cogs(13), gross income(15), rating(16)

unit_price = np.array([float(row[6]) for row in data])
quantity = np.array([int(row[7]) for row in data])
tax = np.array([float(row[8]) for row in data])
total = np.array([float(row[9]) for row in data])
gross_income = np.array([float(row[15]) for row in data])
rating = np.array([float(row[16]) for row in data])

# ARRAY KATEGORIKAL

branch = np.array([row[1] for row in data])
city = np.array([row[2] for row in data])
customer_type = np.array([row[3]for row in data])
gender = np.array([row[4] for row in data])
product_line = np.array([row[5] for row in data])
payment = np.array([row[12] for row in data])
```

AGREGATION BY CATEGORY

ANALISIS PER BRANCH

```python
unique_branches = np.unique(branch)
for b in unique_branches:
    mask = branch == b
    print(f"    Branch {b}  :")
    print(f"    Transaksi   : {mask.sum():4d}")
    print(f"    Revenue     : ${total[mask].sum():>12,.2f}")
    print(f"    Avg rating  : {rating[mask].mean():>6.2f}")

```

ANALISA PER GENDER

```python
unique_gender = np.unique(gender)
for g in unique_gender:
    mask = gender == g
    pct = mask.sum() / len(gender) * 100
    print(f"    Gender {g}  : ")
    print(f"    Transaksi   : ${mask.sum():4d} ({pct:.1f}%)")
    print(f"    Avg spend   : ${total[mask].mean():,.2f}")
```

ANALYSIS PER PAYMENT

```python
unique_payment = np.unique(payment)
for p in payment:
    mask = payment == p
    pct = (mask.sum() / len(payment)) * 100
    rev = total[mask].sum()
    print(f"    {p:15s}: {mask.sum():4d} transaksi ({pct:5.1f}%) | Revenue : ${rev:>12,.2f}")
```

# TOP 3 PRODUCT LINES

```python
unique_products = np.unique(product_line)
product_revenues = [(p, total[product_line == p].sum()) for p in unique_products]
product_revenues.sort(key=lambda x: x[1], reverse=True)
for i, (prod, rev) in enumerate(product_revenues[:3], 1):
    count = (product_line == prod).sum()
    print(f"    {i}. {prod:30s}: ${rev:>12,.2f} ({count} transaksi)")
```

# ADVANCED NUMPY - PERCENTILES DAN CORRELATION

```python
percentiles = np.percentile(total, [25,50,75,90,95])
print(f"    25th percentile (Q1) : ${percentiles[0]:,.2f}")
print(f"    50th percentile (Q2) : ${percentiles[1]:,.2f}")
print(f"    75th percentile (Q3) : ${percentiles[2]:,.2f}")
print(f"    90th percentile      : ${percentiles[3]:,.2f}")
print(f"    95th percentile      : ${percentiles[4]:,.2f}")
```

# IQR ( INTERQUARTILE RANGE )

```python
iqr = percentiles[2] - percentiles[0]
print(f"    QI - Q2     : ${iqr:,.2f}")
```

# CORRELATION ANTARA QUANTITY DAN TOTAL

```python
corr_coef = np.corrcoef(quantity, total)[0, 1]
print("KORELASI:")
print(f"    Quantity vs Total   : {corr_coef:4f} {'(Strong)' if abs(corr_coef) > 0.7 else'(Moderate'}")
```

# ARRAY OPERATION

```python
# TOP 5 TRANSAKSI TERBESAR
top_5_idx = np.argsort(total)[-5:][::-1]
print("TRANKSASI TERBESAR : ")
for i, idx in enumerate(top_5_idx,1):
    print(f"    {i}. ${total[idx]:>8,.2f} | {product_line[idx]:30s} | branch {branch[idx]} | {gender[idx]}")
```

# top 5 rating tertingi


```python
high_rating_mask = rating >= 9.5
high_rating_idx = np.where(high_rating_mask)[0]
high_rating_sorted = high_rating_idx[np.argsort(total[high_rating_mask])[::-1]]
print("TOP 5 HIGH RATING TERTINGGI ( >= 9.5 DENGAN TOTAL TERBESAR : ")
for i, idx in enumerate(high_rating_sorted[:5],1):
    print(f"    {i}. Rating {rating[idx]:.1f} | ${total[idx]:>8,.2f} | {customer_type[idx]:10s} | {payment[idx]}")
```

# RESHAPE MATRIX OPERATIONS

```python
n_samples = 10
feature_matrix = np.column_stack((
    unit_price[:n_samples],
    quantity[:n_samples],
    total[:n_samples],
    rating[:n_samples],
))
```

```python
print(f"FEATURE MATRIX ( SAMPLE {n_samples} ROW X 4 KOLOM) :")
print("     UNIT PRICE  |   QUANTITY    |   TOTAL   |   RATING")
print("-" *60)
for row in feature_matrix:
    print(f"  {row[0]:>10.2f}     |   {row[1]:>8.0f} |   {row[2]:>8.2f}  |   {row[3]:>6.1f}")
```

```python
print("STATISTIK PER KOLOM:")
print(f"    Mean    : {feature_matrix.mean(axis=0)}")
print(f"    Max     : {feature_matrix.max(axis=0)}")
print(f"    Std     : {feature_matrix.std(axis=0)}")
```

# CONDITIONAL OPERATION & WHERE

```python
# KATEGORI PELANGGAN BERDASARKAN SPENDING
spending_category = np.where( total >= 500, 'HIGH SPENDER',
                              np.where(total >= 200, 'Medium Spender', 'low spender'))

unique_cat , counts = np.unique(spending_category, return_counts=True)
```

```python
print(f"KATEGORI PELANGGAN BERDASARKAN SPENDING:")
for cat, count in zip(unique_cat,counts):
    pct = (count / len(spending_category)) * 100
    avg_spent = total[spending_category == cat].mean()
    print(f"    {cat:15s}: {count:4d}  ({pct:5.1f}%)   |   Avg: ${avg_spent:>8,.2f}")
```

# RATING KATEGORI

```python
rating_category = np.where(rating >= 8.0, 'Excellent',
                           np.where(rating >= 6.0, 'Good', 'Poor'))
unique_rat, counts_rat = np.unique(rating_category, return_counts=True)
```

```python
print("KATEGORI RATING:")
for cat, count in zip(unique_rat,counts_rat):
    pct = (count / len(rating_category)) * 100
    print(f"    {cat:10s}:  {count:4d}  transaksi   ({pct:5.1f})")
```

# BUSSINESS INSIGHT

```python
# total REVENUE DAN PROFIT
total_revenue = total.sum()
total_profit = gross_income.sum()
profit_percentage = (total_profit / total_revenue) * 100
```

```python
print(f"    Total Transaksi     : {len(total):,}")
print(f"    Total Revenue       : ${total_revenue:,.2f}  ")
print(f"    Total Profit        : ${total_profit:,.2f}  ")
print(f"    Total Margin        : ${profit_percentage:,.2f}")
print(f"    Avg Transaction     : ${total.mean():,.2f}")
print(f"    Customer Rating     : {rating.mean():,.2f}/10")
```

```python
# best performing branch
print("PERFORMA TERBAIK")
branch_revenues = [(b, total[branch == b].sum()) for b in unique_branches]
best_branch, best_revenue = max(branch_revenues, key=lambda x: x[1])
print(f"    Best Branch     : {best_branch}")
print(f"    Revenue         : ${best_revenue:,.2f}  ")
```

MOST POPULEAR PRODUCT LINE

```python
product_counts = [(p, (product_line == p).sum()) for p in np.unique(product_line)]
best_product, best_count = max(product_counts, key=lambda x: x[1])
print(f"    TOP PRODUCT LINE    : {best_product}")
print(f"    TRANSACTION         : {best_count}  ")
```

BEST USTIOMER SEGMENT

```python
best_gender = max(unique_gender, key=lambda g: total[gender == g].sum())
print(f"    TOP GENDER SEGMENT: {best_gender}")
print(f"    REVENUE           : ${total[ gender == best_gender].sum():,.2f}")
```
