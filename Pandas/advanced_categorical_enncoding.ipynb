# ADVANCED CATEGORICAL ENCODING

Tujuan : Mengubah data kategori ( teks ) menjadi angka yang bisa dibaca mdoel
kapan di pakai  : wajib! model ml hanya bisa baca angka bukan teks

kenapa ini sangat penting?
tanpa encoding:
model error : "cannot convert string to float : "jakarta"

dengan encoding:
- Jakarta [1,0,0]
- Bandung [0,1,0]
- surabaya [0,0,1]
- model bisa belajar pola


metode yang akan di pelajari
- Label encoding - simpel tapi ada pitfall
- one-hot encoding - paling aman dan powerful
- Target encoding - powerfull utnuk data besar
- frequency encoding - cepat dan efektif
- ordinal encoding - untuk data betingkat

IMPACT KE MODEL :
- encoding yang salah -> akurasi turun 10-30%
- encoding yang tepat -> bisa naikkan akurasi 5-15%
